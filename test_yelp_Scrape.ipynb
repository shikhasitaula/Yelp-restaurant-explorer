```python
import requests
from bs4 import BeautifulSoup as soup
from splinter import Browser
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Visit the Mars Facts site
url = 'https://www.yelp.com/search?find_desc=Restaurants&find_loc=San+Jose%2C+CA'
browser.visit(url)
```

```python
# Parse the website
html = browser.html
html_soup = soup(html, 'html.parser')
pprint(html_soup)
```

```python
restaurants = html_soup.find_all('li', class_='lemon--li__373c0__1r9wz border-color--default__373c0__2oFDT')

```

```python
restaurants = html_soup.find_all('li', class_='lemon--li__373c0__1r9wz border-color--default__373c0__2oFDT')
```

```python
restaurant_data = []

for restaurant in restaurants:
    name = restaurant.find('h4').text
    rating = float(restaurant.find('span', class_='display--inline__373c0__2SfH1').div.get('aria-label').split()[0])
    review_count = int(restaurant.find('span', class_='reviewCount__373c0__2r4xT').text.split()[0])
    
    restaurant_data.append({'name': name, 'rating': rating, 'review_count': review_count})
```
