```python
# List of states in the USA
states = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California']  # Add all the states here

# Parameters for the Yelp API request
limit = 50
term = 'restaurant'


listings = []
cols = ['Name', 'Reviews', 'Rating', 'Address', 'Cuisine', 'State']

for state in states:
    search_parameters = {
        'location': state,
        'limit': limit,
        'term': term
       
    }
    
    response = requests.get(url, headers=headers, params=search_parameters)
    raw_data = response.json()

    for business in raw_data['businesses']:
        name = business['name']
        reviews = business['review_count']
        rating = business['rating']
        address = business['location']['display_address'][0]
        cuisine_list = [category['title'] for category in business['categories']]
        cuisine = cuisine_list[0] if cuisine_list else ''
        listings.append([name, reviews, rating, address, cuisine, state])

df = pd.DataFrame.from_records(listings, columns=cols)
print(f'Total Records: {len(df)}')
df.head()
```
